```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
def F(p):
    x = p[0]
    y = p[1]
    z = 4*x*x - 4*x*y + 2*y*y
    return z
pass

def DF(p) :
    x = p[0]
    y = p[1]
    dfx = 8*x -4*y
    dfy = -4*x + 4*y
    return np.array( [dfx, dfy] )
pass
```

```python
delta = 0.025
w = 20
h = 20
x = np.arange(-w, w, delta)
y = np.arange(-h, h, delta)
X, Y = np.meshgrid(x, y)
Z = F([X, Y])
```

```python
fig, ax = plt.subplots()
cs = ax.contour(X, Y, Z)
ax.clabel(cs, inline=1, fontsize=10)
ax.set_title('Contour Plot')
plt.show()
```

```python
def gamma(p1, p0) :
    dp = p1 - p0
    df0 = DF(p0)
    df1 = DF(p1)
    df = df1 - df0
    r = 0.01
    #r = np.transpose( dp ).dot( df )/np.linalg.norm( df, 2 )
    return r, df1
```

```python
r = delta
p = np.array( [ w, h ] )
df = DF( p )

print( f"p={p}, r={r}, df={df}" )

f = F(p)
while f > 0.001 :
    p1 = p - r*df
    r, df = gamma( p1, p )
    p = p1
    f = F(p)
    print( f"p1={p1}, r={r}, df={df}, f={f}" )
```
