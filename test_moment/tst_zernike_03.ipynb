```python
# importing required libraries
import mahotas
import mahotas.demos
from pylab import gray, imshow, show
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
# loading iamge
img = mahotas.demos.load('lena')

# filtering image
img = img.max(2)

shape = img.shape

h = shape[0]
w = shape[1]

radius = sqrt( h*h + w*w )/2

print( f"shape = {shape}, radius = {radius}" )

# showing image
imshow(img)
show()
```

```python
degree = 10 

print( f"shape = {shape}, radius = {radius}, degree = {degree}" )

# computing zernike moments
value = mahotas.features.zernike_moments(img, radius, degree)

# printing value
print( f"zernike moments : \n {value}" )
```
